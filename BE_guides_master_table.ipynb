```python
import numpy as np
import pandas as pd
from datascience import *
    
```

```python
LDLR_file_path = '/Users/allisonjames/Documents/Rotations/Sherwood Lab/LDLR_test_file'
SREBF2_file_path = '/Users/allisonjames/Documents/Rotations/Sherwood Lab/BE screens/SREBF2_table2.txt'
pd.set_option('display.max_columns', None)
```

```python
def get_SA_guides(gene, file_path):
    table = pd.read_csv(file_path, sep='\t')
    entry_1 = table.iloc[0]
    starts = entry_1.loc["exonStarts"].strip(", ").split(',')
    start1 = []
    for i in starts:
        start1 = np.append(start1, int(i)-9))
    start2 = start1-1
    start3 = start1-2
    start4 = start1-3
    start5 = start1-4
    stop1 = start1+19
    stop2 = start2+19
    stop3 = start3+19
    stop4 = start4+19  
    stop5 = start5+19
    guide1 = []
    guide2 = []
    guide3 = []
    guide4 = []
    guide5 = []
    for i in np.arange(len(start1)):
        guide1 = np.append(guide1, "chr19:"+str(int(start1[i]))+"-"+str(int(stop1[i])))
        guide2 = np.append(guide2, "chr19:"+str(int(start2[i]))+"-"+str(int(stop2[i])))
        guide3 = np.append(guide3, "chr19:"+str(int(start3[i]))+"-"+str(int(stop3[i])))
        guide4 = np.append(guide4, "chr19:"+str(int(start4[i]))+"-"+str(int(stop4[i])))
        guide5 = np.append(guide5, "chr19:"+str(int(start5[i]))+"-"+str(int(stop5[i])))
    
    guide_positions = pd.DataFrame({"gene": gene, "site": "splice acceptor", "strand":"+", "guide1":guide1 , "guide2":guide2, "guide3":guide3, "guide4":guide4, "guide5":guide5})
    return guide_positions

def get_SD_guides(gene, file_path):
    table = pd.read_csv(file_path, sep='\t')
    entry_1 = table.iloc[0]
    starts = entry_1.loc["chromStarts"].strip(", ").split(',')
    start1 = []
    for i in np.arange(len(starts)):
        start1 = np.append(start1, int(entry_1.loc["chromStart"]+int(starts[i])+int(entry_1.loc["blockSizes"].split(',')[i])-11))
    start2 = start1-1
    start3 = start1-2
    start4 = start1-3
    start5 = start1-4
    stop1 = start1+19
    stop2 = start2+19
    stop3 = start3+19
    stop4 = start4+19  
    stop5 = start5+19
    guide1 = []
    guide2 = []
    guide3 = []
    guide4 = []
    guide5 = []
    for i in np.arange(len(start1)):
        guide1 = np.append(guide1, "chr19:"+str(int(start1[i]))+"-"+str(int(stop1[i])))
        guide2 = np.append(guide2, "chr19:"+str(int(start2[i]))+"-"+str(int(stop2[i])))
        guide3 = np.append(guide3, "chr19:"+str(int(start3[i]))+"-"+str(int(stop3[i])))
        guide4 = np.append(guide4, "chr19:"+str(int(start4[i]))+"-"+str(int(stop4[i])))
        guide5 = np.append(guide5, "chr19:"+str(int(start5[i]))+"-"+str(int(stop5[i])))
    
    guide_positions = pd.DataFrame({"gene": gene, "site": "splice donor", "strand":"-", "guide1":guide1 , "guide2":guide2, "guide3":guide3, "guide4":guide4, "guide5":guide5})
    return guide_positions    

```

```python
def get_guides(gene, file_path):
    SA_guides = get_SA_guides(gene, file_path)
    SD_guides = get_SD_guides(gene, file_path)
    master_table = SA_guides.append(SD_guides)
    return master_table
```
