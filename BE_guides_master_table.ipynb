```python
def get_3_prime_guides(gene, file_path):
    table = pd.read_csv(file_path, sep='\t')
    entry_1 = table.iloc[0]
    starts = entry_1.loc["exonStarts"].strip(", ").split(',')
    start1 = []
    for i in starts:
        start1 = np.append(start1, int(i)-8)
    if entry_1.loc["strand"] == "+":
        site = "splice acceptor"
    else: 
        site = "splice donor"
    start2 = start1+1
    start3 = start1+2
    start4 = start1+3
    start5 = start1+4
    stop1 = start1+19
    stop2 = start2+19
    stop3 = start3+19
    stop4 = start4+19  
    stop5 = start5+19
    guide1 = []
    guide2 = []
    guide3 = []
    guide4 = []
    guide5 = []
    for i in np.arange(len(start1)):
        guide1 = np.append(guide1, entry_1["chrom"]+":"+str(int(start1[i]))+"-"+str(int(stop1[i])))
        guide2 = np.append(guide2, entry_1["chrom"]+":"+str(int(start2[i]))+"-"+str(int(stop2[i])))
        guide3 = np.append(guide3, entry_1["chrom"]+":"+str(int(start3[i]))+"-"+str(int(stop3[i])))
        guide4 = np.append(guide4, entry_1["chrom"]+":"+str(int(start4[i]))+"-"+str(int(stop4[i])))
        guide5 = np.append(guide5, entry_1["chrom"]+":"+str(int(start5[i]))+"-"+str(int(stop5[i])))
    
    guide_positions = pd.DataFrame({"gene": gene, "site": site, "strand":entry_1.loc["strand"], "chromosome":entry_1["chrom"].strip("chr"),"BE": "ABE", "guide1":guide1 , "guide2":guide2, "guide3":guide3, "guide4":guide4, "guide5":guide5})
    return guide_positions

def get_5_prime_guides(gene, file_path):
    table = pd.read_csv(file_path, sep='\t')
    entry_1 = table.iloc[0]
    ends = entry_1.loc["exonEnds"].strip(", ").split(',')
    start1 = []
    if entry_1.loc["strand"] == "+":
        site = "splice donor"
    else: 
        site = "splice acceptor"
    for i in ends:
        start1 = np.append(start1, int(i)-10)
    start2 = start1-1
    start3 = start1-2
    start4 = start1-3
    start5 = start1-4
    stop1 = start1+19
    stop2 = start2+19
    stop3 = start3+19
    stop4 = start4+19  
    stop5 = start5+19
    guide1 = []
    guide2 = []
    guide3 = []
    guide4 = []
    guide5 = []
    for i in np.arange(len(start1)):
        guide1 = np.append(guide1, entry_1["chrom"]+":"+str(int(start1[i]))+"-"+str(int(stop1[i])))
        guide2 = np.append(guide2, entry_1["chrom"]+":"+str(int(start2[i]))+"-"+str(int(stop2[i])))
        guide3 = np.append(guide3, entry_1["chrom"]+":"+str(int(start3[i]))+"-"+str(int(stop3[i])))
        guide4 = np.append(guide4, entry_1["chrom"]+":"+str(int(start4[i]))+"-"+str(int(stop4[i])))
        guide5 = np.append(guide5, entry_1["chrom"]+":"+str(int(start5[i]))+"-"+str(int(stop5[i])))
    
    guide_positions = pd.DataFrame({"gene": gene, "site": site, "strand":entry_1.loc["strand"], "chromosome":entry_1["chrom"].strip("chr"), "BE": "ABE", "guide1":guide1 , "guide2":guide2, "guide3":guide3, "guide4":guide4, "guide5":guide5})
    return guide_positions

```

```python
def get_guides(gene, file_path):
    three_prime_guides = get_3_prime_guides(gene, file_path)
    five_prime_guides = get_5_prime_guides(gene, file_path)
    master_table = three_prime_guides.append(five_prime_guides)
    return master_table
```
